# Find the Difference

You are given two strings s and t.

String t is generated by random shuffling string s and then add one more letter at a random position.

Return the letter that was added to t.

```python
def findTheDifference( s, t):
    dic={}
    for latter in s:
       if latter not in dic:
           dic[latter]=1
       else:
           dic[latter]=dic[latter]+1
    for l in t:
        if l not in dic:
            return l
        else:
            dic[l]=dic[l]-1
            if dic[l]<0:
                return l
    #return dic#[k for k, v in dic.items() if v == 1][0]
```

```python
s = "a"
t = "aa"

findTheDifference( s, t)
```
